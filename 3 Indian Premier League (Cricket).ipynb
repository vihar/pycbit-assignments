```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/ipl_data.csv")
```

```python
srh = df[df["batting_team"] == "Sunrisers Hyderabad"]
first_inn_runs = srh[srh["inning"] == 1]
second_inn_runs = srh[srh["inning"] == 2]
xlabels = ["batting first", "batting second"]
ylabels = [sum(list(first_inn_runs["total_runs"])), sum(list(second_inn_runs["total_runs"]))]
plt.bar(xlabels, ylabels)
plt.title("Comparision of runs scored by SRH in 1st and 2nd innings")
#shows SRH has scored more runs while batting first
```
