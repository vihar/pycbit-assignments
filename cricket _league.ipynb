```python
srh = df[df["winner"] == "Royal Challengers Bangalore"]
winner_in_2017 = srh[srh["season"] == 2017]
winner_in_2009 = srh[srh["season"] == 2009]
xlabels = ["2017 season", "2009 season"]
ylabels = [len(winner_in_2017), len(winner_in_2009)]
plt.bar(xlabels, ylabels)
plt.title("Counting the winnig rate of teams in particular seasons")
#shows SRH has scored more runs while batting first
```
